```python
name = input("Enter student name: ")

maths = int(input("Maths marks: "))
science = int(input("Science marks: "))
english = int(input("English marks: "))

total = maths + science + english

percentage = total / 3

if percentage >= 90:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 50:
    grade = "C"
else:
    grade = "Fail"

print("\n----- RESULT -----")
print("Student:", name)
print("Total:", total)
print("Percentage:", percentage)
print("Grade:", grade)
```
