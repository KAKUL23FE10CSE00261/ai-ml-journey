```python
to_do_list = []

def add_task():
    task = input("Enter a task: ")
    to_do_list.append(task)
    print("Task Added Successfully!")

def view_task():
    if len(to_do_list) == 0:
        print("No Task Found.")
        return
    
    print("\n----- TO DO LIST -----")

    for task in to_do_list:
        print(task)
    
def remove_task():
    delete_task = input("Enter a task to remove it: ")

    for task in to_do_list:
        if task == delete_task:
            to_do_list.remove(task)

            print("succesfully Deleted!!")
            return
        
    print("Task Not Found")

# ---------------- MAIN PROGRAM ----------------
while True:

    print("\n===== TO DO LIST MENU =====")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Remove Task")
    print("4. Exit")

    choice = input("Enter your choice: ")

    # ADD TASK
    if choice == "1":
        add_task()

    # VIEW TASKS
    elif choice == "2":
        view_task()

    # REMOVE TASK
    elif choice == "3":
        remove_task()

    # EXIT
    elif choice == "4":
        print("Program Exited.")
        break

    else:
        print("Invalid Choice")

            
```
