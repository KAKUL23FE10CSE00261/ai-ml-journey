Create BankAccount class

Constructor:
    initialize balance

deposit():
    add money

withdraw():
    check balance
    subtract money

show_balance():
    print balance

```python
class BankAccount():

    def __init__(self, balance):
        self.balance = balance

    def deposiit(self, amount):
        self.amount = amount

        self.balance += amount
        print(f"₹{amount} deposited successfully.")
    
    def withdraw(self, withdraw_money):
        self.withdraw_money = withdraw_money

        if self.balance < withdraw_money:
            print("Insufficient money")

        else:
            self.balance -= withdraw_money
            print(f"₹{withdraw_money} withdrawn successfully.")

    def showBalance(self):
        print(f"Current Balance: ₹{self.balance}")

account1 = BankAccount(10000)
account2 = BankAccount(1089000)

account1.deposiit(5677080)
account1.withdraw(754678)
account1.showBalance()


            



```
