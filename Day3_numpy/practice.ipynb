#### Task 16 — Student Marks Analyzer

Create NumPy array of student marks.

Find:

- average marks
- topper marks
- failed students
- percentage above 75

```python
# Student marks
marks = np.array([78, 45, 90, 67, 34, 88, 76, 55, 92, 40])

# 1. Average marks
average_marks = np.mean(marks)
print("Average Marks:", average_marks)

# 2. Topper Marks
topper_marks = np.max(marks)
print("Topper Marks:", topper_marks)

# 3. Failed Students
failed_students = np.min(marks)
print("Failed students:", failed_students)

#4. Percentage above 75%
above_75 = marks[marks>75]
print("above 75%:", above_75)
```

#### Task 18 — Manual Matrix Multiplication

Implement matrix multiplication:

using loops ONLY

Then compare with:

np.dot()

```python
# Matrices
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

# Dimensions
rows_A = len(A)
cols_A = len(A[0])

rows_B = len(B)
cols_B = len(B[0])

result = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

for i in range(rows_A):
    for j in range(cols_B):
        for k in range(cols_A):
            result[i][j] += A[i][k] *  B[k][j]

print("Manual MAtrix Multiplication: ")
for row in result:
    print(row)

numpy_result = np.dot(A,B)

print("Using np.dot()")
print(numpy_result)
```
